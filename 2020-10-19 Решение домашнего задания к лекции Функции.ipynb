```python
import copy

documents = [ {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, 
             {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, 
             {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} ] 

directories = { '1': ['2207 876234', '11-2'], 
               '2': ['10006'], '3': [] }


def owner_by_number():
    input_number = input('Введите номер документа: ')
    for data in documents:
        if input_number == data['number']:
            print('Владелец документа' ,data['name'])
            return
    print('Документ не найден в базе')
    
    
def shelf_by_number():
    input_number = input('Введите номер документа: ')
    for shelf, numb in directories.items():
        for digit in numb:
            if digit == input_number:
                print('Документ хранится на полке:', shelf)
                return
    print('Документ не найден в базе')
    

def all_info():    
    for data in documents:             
        for shelf, numb in directories.items():
            if data['number'] in numb:
                print(f'№ {data["number"]}, тип: {data["type"]}, владелец: {data["name"]}, полка хранения: {shelf}')

                
def add_shelf():
    shelf_number_input = input('Введите номер полки:')
    if shelf_number_input not in directories.keys():
        directories[shelf_number_input] = []
        string = directories.keys()
        cleared_string = ''
        for i in string:
            cleared_string += i + ', '
            correct_cleared_string = cleared_string.strip(', ')
        print('Полка добавлена. Текущий перечень полок: ', correct_cleared_string)
    else:
        print('Такая полка уже существует. Текущий перечень полок: ', ', '.join(list(directories.keys())))


def delete_shelf():
    directories2 = copy.deepcopy(directories)
    # здесь происходит пользовательский ввод номера полки и преобразование в строковый тип функцией str()
    input_empty_shelf = str(input('Введите номер полки:'))
    # здесь мы получаем содержимое словаря по ключу input_empty_shelf, полученному на предыдущем шаге
    shelf_content = directories2.get(input_empty_shelf)
    # если список пустой
    if shelf_content == []:
        del directories[input_empty_shelf]
        print('Полка удалена. Текущий перечень полок:', list(directories.keys()))
        return
    # если такой полки нет
    if shelf_content == None:
        print('Такой полки нет')
        return
    # если на полке есть какие-либо документы
    if shelf_content:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', list(directories.keys()))          
    return
    
    

def main():
    while True:
        input_command = input('Введите команду:')
        if input_command == 'q':
            print('Программа завершена')
            break
        elif input_command == 'p':
            owner_by_number()
        elif input_command == 's':
            shelf_by_number()
        elif input_command == 'i':
            all_info()
        elif input_command == 'ads':
            add_shelf()
        elif input_command == 'ds':
            delete_shelf()            
        else:
            print('Неверный ввод')
            
            
main()
```
